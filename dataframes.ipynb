1. Copy the code from the lesson to create a dataframe
full of student grades.


```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
df
```

```python
# 1a. Create a column named passing_english that indicates
# whether each student has a passing grade in english.
df['passing_english'] = df.english > 70
```

```python
df
```

```python
# 1b. Sort the english grades by the passing_english column.
# How are duplicates handled? -by index
df.sort_values(by='passing_english')
```

```python
# 1c. Sort the english grades first by passing_english and then
# by student name. All the students that are failing english
# should be first, and within the students that are failing
# english they should be ordered alphabetically. The same
# should be true for the students passing english. (Hint:
# you can pass a list to the .sort_values method)
df.sort_values(by=['passing_english','name'])
```

```python
# 1d. Sort the english grades first by passing_english, and
# then by the actual english grade, similar to how we did
# in the last step.
df.sort_values(by=['passing_english','english'])
```

```python
# 1e. Calculate each students overall grade and add it as a
# column on the dataframe. The overall grade is the average
# of the math, english, and reading grades.
df['overall_grade'] = round((df.math + df.english + df.reading)/3, 2)
df
```

```python
# Rename the cty column to city.
mpg.rename(columns={'cty':'city'})
```

```python
# Rename the hwy column to highway.
mpg.rename(columns={'hwy':'highway'})
```

```python
# Do any cars have better city mileage than highway mileage?
mpg[mpg.cty>mpg.hwy]
```

```python
# Create a column named mileage_difference this column
# should contain the difference between highway and city
# mileage for each car.
mpg['mileage_difference'] = mpg.hwy - mpg.cty
mpg
```

```python
# Which car (or cars) has the highest mileage difference?
mpg[mpg.mileage_difference==mpg.mileage_difference.max()]
```

```python
# Which compact class car has the lowest highway mileage?
mpg_compact = mpg[mpg['class']=='compact']
mpg_compact[mpg_compact.hwy==mpg_compact.hwy.min()]
```

```python
# The best?
mpg_compact[mpg_compact.hwy==mpg_compact.hwy.max()]
```

```python
# Create a column named average_mileage that is the mean
# of the city and highway mileage.
mpg['average_mileage'] = (mpg.cty + mpg.hwy)/2
mpg
```

```python
# Which dodge car has the best average mileage?
mpg_dodge = mpg[mpg.manufacturer=='dodge']
mpg_dodge[mpg_dodge.average_mileage==mpg_dodge.average_mileage.max()]
```

```python
# The worst?
mpg_dodge[mpg_dodge.average_mileage==mpg_dodge.average_mileage.min()]
```
