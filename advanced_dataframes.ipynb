```python
# Copy the users and roles DataFrames from the examples above.
# Create the users DataFrame.
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# What is the result of using a right join on the DataFrames?
users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)
```

```python
# What is the result of using an outer join on the DataFrames?
users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
```

```python
# What happens if you drop the foreign keys from the
# DataFrames and try to merge them?
# users.drop(columns=['role_id']).merge(roles.drop(columns=['id']), how='outer', indicator=True)
users.drop(columns=['role_id']).merge(roles, how='outer', indicator=True)
```

```python
# Create a column named mileage_difference like you did in
# the DataFrames exercises; this column should contain the
# difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.hwy - mpg.cty
```

```python
# Create a column named average_mileage like you did in the
# DataFrames exercises; this is the mean of the city and
# highway mileage.
mpg['average_mileage'] = (mpg.hwy + mpg.cty)/2
```

```python
# Create a new column on the mpg dataset named is_automatic
# that holds boolean values denoting whether the car has
# an automatic transmission.
# mpg['is_automatic'] = mpg['trans'].str.contains('auto')
mpg['is_automatic'] = np.where(mpg['trans'].str.contains('auto'),'auto','manual')
mpg.is_automatic
```

```python
# Using the mpg dataset, find out which which manufacturer
# has the best miles per gallon on average?
mnf_mpg = mpg.groupby('manufacturer').average_mileage.agg(['mean'])
mnf_mpg.nlargest(1, ['mean'])
```

```python
# Do automatic or manual cars have better miles per gallon?
tr_mpg = mpg.groupby('is_automatic').average_mileage.agg(['mean'])
tr_mpg
```

```python
# Use your get_db_url function to help you explore the
# data from the chipotle database.
chip_url = get_db_url('chipotle')
chipotle = pd.read_sql('select * from orders', chip_url)
```

```python
chipotle.info()
```

```python
chipotle.item_price = chipotle.item_price.str.strip('$').astype(float)
```

```python
chipotle.item_price
```

```python
# What is the total price for each order?
chipotle.groupby('order_id').item_price.agg(['mean'])
```

```python
# What are the most popular 3 items?
chip_items = chipotle.groupby('item_name').quantity.agg('sum')
chip_items.nlargest(3)
```

```python
# Which item has produced the most revenue?
chipotle['revenue'] = chipotle.quantity * chipotle.item_price
chip_rev = chipotle.groupby('item_name').revenue.agg(['sum'])
chip_rev.nlargest(1,'sum')
```

```python
# Join the employees and titles DataFrames together.
emp_title = employees.merge(titles, left_on='emp_no', right_on='emp_no')
emp_title.sample(5)
```

```python
# For each title, find the hire date of the employee that
# was hired most recently with that title.
emp_title.groupby('title').hire_date.agg('max')
```

```python
# Write the code necessary to create a cross tabulation of
# the number of titles by department. (Hint: this will
# involve a combination of SQL code to pull the necessary
# data and python/pandas code to perform the manipulations.)
dept_title = pd.read_sql('''
select title, dept_name
from dept_emp
join departments
	using(dept_no)
join titles
	using(emp_no)
''', emp_url)
```

```python
pd.crosstab(dept_title.dept_name, dept_title.title)
```
