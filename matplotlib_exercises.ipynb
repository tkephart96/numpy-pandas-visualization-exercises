# matplotlib exercises

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Use matplotlib to plot the following equation:
# y=x^2−x+2
# You'll need to write the code that generates the
# x and y points.
# Add an annotation for the point 0, 0, the origin.
x = list(range(-49,50))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(-4.5, 300), arrowprops={'facecolor':'blue'})
plt.title('$y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$',rotation=0)
plt.show()
```

```python
# Create and label 4 separate charts for the following
# equations (choose a range for x that makes sense):
# y=√x
# y=x^3
# y=2^x
# y=1/(x+1)
# You can use functions from the math module to help implement
# some of the equations above.
x = np.linspace(-5,6)
y1 = [np.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

# y=√x
# first subplot
plt.subplot(221)
plt.plot(x, y1)
plt.title('$y = \sqrt{x}$')

plt.show()
```

```python
# y=x^3
# second subplot
plt.subplot(222)
plt.plot(x, y2)
plt.title('$y = x^3$')

plt.show()
```

```python
# y=2^x
# third subplot
plt.subplot(223)
plt.plot(x, y3)
plt.title('$y = 2^x$')

plt.show()
```

```python
# y=1/(x+1)
# fourth subplot
plt.subplot(224)
plt.plot(x, y4)
plt.title('$y = 1 / (x + 1)$')

plt.show()
```

```python
# Combine the figures you created in the last step into one
# large figure with 4 subplots.
x = np.linspace(-5,6,100)
y1 = [np.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.figure(figsize=(10,7))

# first subplot
plt.subplot(221)
plt.plot(x, y1)
plt.title('$y = \sqrt{x}$')

# second subplot
plt.subplot(222)
plt.plot(x, y2)
plt.title('$y = x^3$')

# third subplot
plt.subplot(223)
plt.plot(x, y3)
plt.title('$y = 2^x$')

# fourth subplot
plt.subplot(224)
plt.plot(x, y4)
plt.title('$y = 1 / (x + 1)$')

plt.suptitle('4 Equations')
plt.show()
```

```python
# Combine the figures you created in the last step into one
# figure where each of the 4 equations has a different color
# for the points. Be sure to include a legend and an
# appropriate title for the figure.
x = np.linspace(-5,6,100)
y1 = [np.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

# first plot
plt.plot(x, y1, c='blue', label='$y = \sqrt{x}$')

# second plot
plt.plot(x, y2, c='green', label='$y = x^3$')

# third plot
plt.plot(x, y3, c='red', label='$y = 2^x$')

# fourth plot
plt.plot(x, y4, c='orange', label='$y = 1 / (x + 1)$')

plt.ylim(-10,10)
plt.legend(loc='upper left')
plt.title('4 Equations')
plt.show()
```
